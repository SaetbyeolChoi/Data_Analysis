**TO DO:** Write 3-5 sentences about TSA.
> *  Time series analysis (TSA) involves examining data collected over time to discover patterns such as trends, seasonality, and residuals, which are noise. Trends represent long-term increases or decreases, while seasonality reflects patterns that repeat at fixed intervals, such as monthly or yearly cycles. Residuals capture random or unpredictable changes in the data. Breaking down time series data into these components helps us understand the underlying behavior and build better forecasting models. As a result, identifying patterns through time series data analysis is important for informing decision making and predicting future outcomes.






**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/Projects(TDSP)/Motor_Vehicle_Collisions_-_Crashes_20250108.csv"
data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x=average_crashes_per_hour.index, y=average_crashes_per_hour.values)
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby('CRASH DATE').size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Date')
plt.ylabel('Number of Crashes')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonality')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

**Step 3:** Next, we will begin "Severity Mapping." We'll plot crashes on the map and code them based on severity, distinguishing between crashes that resulted in injuries and those that led to fatalities. This will give us a visual representation of where severe crashes tend to occur. </br>

You may initially want to code these incidents by using different colors, but it's important to think about your map and how accessible it is. Will a color-coded map be easy to read for everyone? Let's make a map that's inclusive for people with color blindness by also creating differently-shaped markers (squares, circles, and triangles) for crashes, injuries, and fatalities.

````python
#TODO: Continue building your heatmap
# Sample a subset of the data for visualization
sample_data_severity = data_geo.sample(n=1000, random_state=42)

# Create a base map
m_severity = folium.Map(location=[40.730610, -73.935242], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in sample_data_severity.iterrows():
    if row['NUMBER OF PERSONS KILLED'] > 0:
        color = "red"  # Fatalities

        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=3,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


    elif row['NUMBER OF PERSONS INJURED'] > 0:
        color = "orange"  # Injuries
        folium.CircleMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          radius=5,
          color=color,
          fill=True,
          fill_color=color
       ).add_to(m_severity)
    else:
        color = "green"  # No injuries or fatalities
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


m_severity.save("severity.html")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Load datasets
collisions = pd.read_csv("/content/drive/MyDrive/Projects(TDSP)/Motor_Vehicle_Collisions_-_Crashes_20250108.csv")
intersections = pd.read_csv("/content/drive/MyDrive/Projects(TDSP)/VZV_Priority_Intersections_20250123.csv")
````

````python
# Display missing values in both datasets
print("Missing Values in Motor Vehicle Collisions Dataset:")
print(collisions.isnull().sum() / len(collisions))
print("\nMissing Values in Priority Intersections Dataset:")
print(intersections.isnull().sum())
````

````python
# Clean dataset
collisions_cleaned = collisions.dropna(subset=['LATITUDE', 'LONGITUDE', 'CRASH DATE', 'CRASH TIME']).copy()
````

````python
# Convert CRASH DATE and CRASH TIME to datetime
collisions_cleaned['CRASH DATE'] = pd.to_datetime(collisions_cleaned['CRASH DATE'], errors='coerce')
collisions_cleaned['CRASH TIME'] = pd.to_datetime(collisions_cleaned['CRASH TIME'], format='%H:%M', errors='coerce')
````

````python
# Combine CRASH DATE and CRASH TIME into a single datetime column
collisions_cleaned['CRASH_DATETIME'] = pd.to_datetime(
    collisions_cleaned['CRASH DATE'].dt.strftime('%Y-%m-%d') + ' ' +
    collisions_cleaned['CRASH TIME'].dt.strftime('%H:%M:%S'),
    format='%Y-%m-%d %H:%M:%S',
    errors='coerce'
)
````

````python
# Drop unnecessary columns and fill missing values
collisions_cleaned = collisions_cleaned.drop(columns=['LOCATION', 'OFF STREET NAME'], errors='ignore')
collisions_cleaned['BOROUGH'] = collisions_cleaned['BOROUGH'].fillna('Unknown')
````

````python
# Round to 5 decimal places for better matching
collisions_cleaned['LATITUDE'] = collisions_cleaned['LATITUDE'].round(5)
collisions_cleaned['LONGITUDE'] = collisions_cleaned['LONGITUDE'].round(5)
````

````python
# Filter for pedestrian related crashes
pedestrian_crashes = collisions_cleaned[
    (collisions_cleaned['NUMBER OF PEDESTRIANS INJURED'] > 0) |
    (collisions_cleaned['NUMBER OF PEDESTRIANS KILLED'] > 0)
]

print(f"\nTotal pedestrian-related crashes: {len(pedestrian_crashes)}")
````

````python
# Clean dataset
intersections_cleaned = intersections.drop(columns=['X', 'Y'], errors='ignore')
intersections_cleaned['LAT'] = pd.to_numeric(intersections_cleaned['LAT'], errors='coerce').round(5)
intersections_cleaned['LONG'] = pd.to_numeric(intersections_cleaned['LONG'], errors='coerce').round(5)
intersections_cleaned = intersections_cleaned.dropna(subset=['LAT', 'LONG'])

print(f"Total priority intersections: {len(intersections_cleaned)}")
````

````python
# Merge datasets
intersection_crashes = pd.merge(
    pedestrian_crashes,
    intersections_cleaned,
    left_on=['LATITUDE', 'LONGITUDE'],
    right_on=['LAT', 'LONG'],
    how='inner'
)

print(f"Total pedestrian-related crashes at priority intersections: {len(intersection_crashes)}")
````

````python
# Analysis 1: Distribution of crash severity by borough
borough_crashes = pedestrian_crashes.groupby('BOROUGH')[['NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED']].sum()

# Plot pedestrian crash severity by borough
borough_crashes.plot(kind='bar', figsize=(12, 6), stacked=True, color=['#1f77b4', '#ff7f0e'])
plt.title('Pedestrian Crash Severity by Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Pedestrian Crashes')
plt.xticks(rotation=45)
plt.legend(["Injuries", "Fatalities"], title="Crash Severity")
plt.tight_layout()
plt.show()
````

Brooklyn had the highest pedestrian crash rate, followed by Queens and Manhattan. Staten Island had the lowest crash rate. To improve safety, infrastructure upgrades in Brooklyn, Queens, and Manhattan, such as advanced pedestrian signals and traffic calming, should be prioritized. In the case of Staten Island, regular monitoring should be maintained to ensure safety standards are maintained.

````python
# Analysis 2: Plot top intersections with the highest crashes
if len(intersection_crashes) > 0:
    top_intersections = intersection_crashes.groupby(['Street1', 'Street2']).size().sort_values(ascending=False).head(10)
    if not top_intersections.empty:
        top_intersections.plot(kind='barh', figsize=(10, 6), color='steelblue')
        plt.title('Top 10 Priority Intersections with Pedestrian Crashes')
        plt.xlabel('Number of Crashes')
        plt.ylabel('Intersection')
        plt.tight_layout()
        plt.show()
    else:
        print("No data available for top intersections.")
else:
    print("No crashes at priority intersections after merging.")
````

Intersections such as Delancy and Essex streets, and Atlantic and Rockaway avenues appear to be the most dangerous areas and should be prioritized for interventions such as traffic calming measures and pedestrian infrastructure improvements.

````python
# Analysis 3: Pedestrian Crashes by Day of the Week
pedestrian_crashes['Day of Week'] = pedestrian_crashes['CRASH_DATETIME'].dt.day_name()
weekly_crash_counts = pedestrian_crashes['Day of Week'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(
    x=weekly_crash_counts.index,
    y=weekly_crash_counts.values,
    palette='cubehelix'
)
plt.title('Pedestrian Crashes by Day of the Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

Friday has the highest pedestrian crashes, potentially due to increased pedestrian and vehicular activity at the end of the workweek. Sunday has the least, possibly because of reduced traffic volumes and pedestrian activity.

````python
# Analysis 4: Day of the Week for Crashes at Priority Intersections
if len(intersection_crashes) > 0:
    intersection_crashes['Day of Week'] = intersection_crashes['CRASH_DATETIME'].dt.day_name()
    crashes_by_day = intersection_crashes['Day of Week'].value_counts()

    plt.figure(figsize=(10, 6))
    sns.barplot(
        x=crashes_by_day.index,
        y=crashes_by_day.values,
        palette="Set2"
    )
    plt.title("Crashes at Priority Intersections by Day of Week")
    plt.xlabel("Day of the Week")
    plt.ylabel("Number of Crashes")
    plt.tight_layout()
    plt.show()
else:
    print("No crashes at priority intersections available for day-of-week analysis.")
````

Friday remains the most dangerous day at priority intersections, mirroring the broader trend in pedestrian crashes, while Sunday has the least.

````python
# Analysis 5: Top Contributing Factors for Crashes at Priority Intersections
if len(intersection_crashes) > 0:
    contributing_factors = (
        intersection_crashes['CONTRIBUTING FACTOR VEHICLE 1']
        .value_counts()
        .head(10)
    )

    plt.figure(figsize=(10, 6))
    sns.barplot(
        y=contributing_factors.index,
        x=contributing_factors.values,
        palette="coolwarm"
    )
    plt.title("Top Contributing Factors in Crashes at Priority Intersections")
    plt.xlabel("Number of Crashes")
    plt.ylabel("Contributing Factor")
    plt.tight_layout()
    plt.show()
else:
    print("No crashes at priority intersections available for analysis.")
````

The most common contributing factors, other than â€œUnspecified,â€ included â€œFailure to Yield Right-of-wayâ€ and â€œDriver Inattention/Distraction,â€ indicating that behavioral issues are an important area for intervention.

````python
# Analysis 6: Crashes at Priority Intersections by Borough
if len(intersection_crashes) > 0:
    crashes_by_borough = intersection_crashes['BOROUGH'].value_counts()

    plt.figure(figsize=(10, 6))
    sns.barplot(
        x=crashes_by_borough.index,
        y=crashes_by_borough.values,
        palette="Blues_d"
    )
    plt.title("Crashes at Priority Intersections by Borough")
    plt.xlabel("Borough")
    plt.ylabel("Number of Crashes")
    plt.tight_layout()
    plt.show()
else:
    print("No crashes at priority intersections available for borough-wise analysis.")
````

Brooklyn and Manhattan have the highest number of crashes at priority intersections, further highlighting the need for borough-level interventions such as improved crosswalks, improved signal timing and awareness campaigns.
